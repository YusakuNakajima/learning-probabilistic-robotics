```python
import sys              ###motion_test_header
sys.path.append('../scripts/')
from robot import *
```

```python
world = World(60.0, 0.1)                    ###motion_test_linear

initial_pose = np.array([0, 0, 0]).T
linear = Agent(0.1, 0.0) 
robots = []
for i in range(100):
    r = Robot(initial_pose, sensor=None, agent=linear) 
    world.append(r)
    robots.append(r)

### アニメーション実行 ###
world.draw()
```

```python
import pandas as pd ###motion_test_stats（下のセルまで。データは上の5行程度を掲載）
poses = pd.DataFrame([ [math.sqrt(r.pose[0]**2 + r.pose[1]**2), r.pose[2]] for r in robots], columns=['r', 'theta'])
poses
```
